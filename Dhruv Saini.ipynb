```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
with open('C:/Users/dhruv/Desktop/Neolen/data.json') as data_file:
    data = json.load(data_file)
data
```

```python
df = pd.io.json.json_normalize(data)
df
```

```python
dups = df.pivot_table(index = ['req.body.note.desc' , 'isSafe'], aggfunc ='size')
d = dups.to_frame() 
d
```

```python
X = df.iloc[: , 2].values
X
```

```python
y = df.iloc[: , 0].values
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['req.body.note.desc'])
X1 = X.toarray()
len(X1)
```

```python
 # Encoding the Dependent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state = 0)
classifier1.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier1.predict(X_test)

# K-Fold cross validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier1, X=X_train, y=y_train, cv=10)
model_accuracy = accuracies.mean()
model_standard_deviation = accuracies.std()
print("model_accuracy" , model_accuracy)

print("Std deviations"  , model_standard_deviation)


# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 500, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print( 'Confusion matrix' , cm)

accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
model_accuracy = accuracies.mean()
model_standard_deviation = accuracies.std()
print("model_accuracy" , model_accuracy)

print("Std deviations"  , model_standard_deviation)

```
